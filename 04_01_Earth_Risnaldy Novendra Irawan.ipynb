## Import Library

```python
import pandas as pd
import numpy as np
```

## Input Dataset

```python
#Membaca data dan memasukkannya ke dalam bentuk Pandas Dataframe
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
df
```

```python
#Jalankan code untuk mengecek nama kolom yang tersedia
df.columns
```

```python
#Hilangkan kolom yang dirasa tidak relevan dengan model (contoh: ID). None dapat diisi dengan nama-nama kolom yang akan digunakan.
#Contoh df = df[['X1','X2', 'Y']].copy()

#START CODE
df1 = df[['Geography','Gender','Age','Tenure','Balance','EstimatedSalary']].copy()
df1
```

## Preprocessing

```python
#Lakukan One-Hot Encoder untuk data categorical, dengan fungsi pandas get_dummies

#START CODE
# df = None

df2 = pd.get_dummies(df1, columns=['Geography','Gender'], drop_first=True)
df2
```

```python
#Pisahkan mana X (feature) dengan Y,
#Y adalah kolom "Exited"

#START CODE
X = df["Age"]
y = df["Exited"]
```

```python
#Lakukan Scaler dan/atau Noermalisasi jika diperlukan
from sklearn.preprocessing import MinMaxScaler

# Inisialisasi scaler
scaler = MinMaxScaler()

# Melakukan scaling pada fitur "EstimatedSalary"
X_transform = scaler.fit_transform(df2[["Age"]])

# Hasil akan berupa array numpy, Anda dapat mengonversinya ke dalam Pandas DataFrame jika diperlukan
X_scaled = pd.DataFrame(X_transform, columns=["Age"])

# Menyimpan data X yang telah di-scaling
df2["Age"] = X_scaled["Age"]

df2
```

```python
#Ini digunakan jika dilakukan scaler/Normalisasi. Jika tidak, code ini bisa dilewat dan diganti dengan code yang ada di dalam komen
# X_transform = pd.DataFrame(X_transform, columns = X.columns)
# #X_transform = X.copy()

# Melakukan scaling pada fitur "EstimatedSalary"
X_transform = scaler.fit_transform(df2[["Age"]])

# Menggantikan nilai X dengan X yang sudah di-transformasi
X_transform = pd.DataFrame(X_transform, columns=["Age"])
```

## Train-Test Split

```python
#Split menjadi train dan test dengan test_size 25%
#Tidak perlu mengubah code ini

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size = 0.25,random_state = 123)
```

# Modeling

## Model1

### Soal :
Jelaskan secara Singkat Model pertama yang digunakan!

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Split menjadi train dan test dengan test_size 25%
# Tidak perlu mengubah code ini
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_transform, y, test_size=0.25, random_state=123)

# Membuat objek model Multinomial Naive Bayes
model = MultinomialNB()

# Membuat daftar hyperparameter yang akan diuji
param_grid = {
    'alpha': [0.1, 0.5, 1.0, 2.0, 5.0],  # Hyperparameter alpha yang akan diuji
}

# Membuat objek GridSearchCV untuk mencari parameter terbaik
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5)

# Melatih model pada data pelatihan dengan menggunakan GridSearchCV
grid_search.fit(X_train, y_train)

# Melihat parameter terbaik yang telah ditemukan
best_params = grid_search.best_params_
print("Parameter terbaik:", best_params)

# Membuat model final dengan parameter terbaik yang ditemukan
best_model = MultinomialNB(alpha=best_params['alpha'])
best_model.fit(X_train, y_train)

# Mengevaluasi model pada data pengujian
y_pred = best_model.predict(X_test)

# Menampilkan classification report dengan penanganan zero division
print("Classification Report:")
print(classification_report(y_test, y_pred, zero_division=0))

# Menampilkan confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Menampilkan akurasi model
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi model di atas data pengujian:", accuracy)
```

## Model2
### Soal :
Jelaskan secara Singkat Model ke-2 yang digunakan!

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Membuat objek model Decision Tree
decision_tree_model = DecisionTreeClassifier()

# Membuat daftar hyperparameter yang akan diuji
param_grid = {
    'criterion': ['gini', 'entropy'],  # Kriteria pemilihan node
    'max_depth': [None, 10, 20, 30, 40],  # Maksimum kedalaman pohon
    'min_samples_split': [2, 5, 10],  # Jumlah sampel minimum untuk membagi node internal
    'min_samples_leaf': [1, 2, 4]  # Jumlah sampel minimum pada node daun
}

# Membuat objek GridSearchCV untuk mencari parameter terbaik
tree_grid_search = GridSearchCV(estimator=decision_tree_model, param_grid=param_grid, cv=5)

# Melatih model pada data pelatihan dengan menggunakan GridSearchCV
tree_grid_search.fit(X_train, y_train)

# Melihat parameter terbaik yang telah ditemukan
best_tree_params = tree_grid_search.best_params_
print("Parameter terbaik untuk Decision Tree:", best_tree_params)

# Membuat model Decision Tree dengan parameter terbaik yang ditemukan
best_tree_model = DecisionTreeClassifier(criterion=best_tree_params['criterion'],
                                         max_depth=best_tree_params['max_depth'],
                                         min_samples_split=best_tree_params['min_samples_split'],
                                         min_samples_leaf=best_tree_params['min_samples_leaf'])
best_tree_model.fit(X_train, y_train)

# Mengevaluasi model pada data pengujian
y_tree_pred = best_tree_model.predict(X_test)

# Menampilkan classification report dengan penanganan zero division
print("Classification Report untuk Decision Tree:")
print(classification_report(y_test, y_tree_pred, zero_division=0))

# Menampilkan confusion matrix
print("Confusion Matrix untuk Decision Tree:")
print(confusion_matrix(y_test, y_tree_pred))

# Menampilkan akurasi model Decision Tree
tree_accuracy = accuracy_score(y_test, y_tree_pred)
print("Akurasi model Decision Tree di atas data pengujian:", tree_accuracy)
```

## Model3
### Soal :
Jelaskan secara Singkat Model ke-3 yang digunakan!

```python
# #Pilih salah satu metode Machine Leaarning

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Inisialisasi model Random Forest
model_rf = RandomForestClassifier()

# Daftar hyperparameter yang akan diuji
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Inisialisasi GridSearchCV
grid = GridSearchCV(estimator=model_rf, param_grid=param_grid, scoring='accuracy', cv=10)

# Melatih model pada berbagai kombinasi hyperparameter
grid.fit(X_train, y_train)

# Mendapatkan hyperparameter terbaik
best_params = grid.best_params_

print("Hyperparameter terbaik:", best_params)

# Membuat model Random Forest dengan hyperparameter terbaik
best_model_rf = RandomForestClassifier(**best_params)

# Melatih model dengan hyperparameter terbaik
best_model_rf.fit(X_train, y_train)

# Melakukan prediksi pada data uji
y_pred_rf = best_model_rf.predict(X_test)

# Evaluasi model
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy_rf = accuracy_score(y_test, y_pred_rf)
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
class_report_rf = classification_report(y_test, y_pred_rf)

print("Akurasi Random Forest:", accuracy_rf)
print("\nConfusion Matrix Random Forest:\n", conf_matrix_rf)
print("\nClassification Report Random Forest:\n", class_report_rf)
```
